## Making Age groups

```python
def age_group(age):
    if 18 <= age <= 28:
        return "18-28"
    elif 29 <= age <= 39:
        return "29-39"
    elif 40 <= age <= 50:
        return "40-50"
    else:
        return "51-60"

data['Age_Group'] = data['Age'].apply(age_group)
```

```python
data.head()
```

## Grouping App Counts

```python
def app_usage_group(count):
    if 1 <= count <= 10:
        return "Light (1-10)"
    elif 11 <= count <= 20:
        return "Moderate (11-20)"
    elif 21 <= count <= 35:
        return "High (21-35)"
    else:
        return "Heavy (36-50)"

data['App_Usage_Count_Group'] = data['App_Usage_Count'].apply(app_usage_group)
```
